```python
# ---------------------------------------------------------------------------
# imports
# ---------------------------------------------------------------------------

from plot3D import plot3D
from xyzData import xyzData
```

```python
# ---------------------------------------------------------------------------
# example
# ---------------------------------------------------------------------------

exampleData = r'SampleDataset.csv'
```

```python
# ---------------------------------------------------------------------------
# read CSV file
# ---------------------------------------------------------------------------

f = xyzData(exampleData)

f.filterIPR(1.0)
f.filterNaN('local magnitude')
```

```python
# ---------------------------------------------------------------------------
# plot
# ---------------------------------------------------------------------------

plot3D(f.current, f.index['local magnitude'])
```
